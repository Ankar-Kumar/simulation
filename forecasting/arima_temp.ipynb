```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas import read_csv
# from datetime import datetime
```

Read the file

```python
df  = pd.read_csv('MaunaLoaDailyTemps.csv')
df.head()
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
df.head()
```

#Drop the unnessary column

```python
drop_col = ['MinTemp', 'MaxTemp', 'Sunrise', 'Sunset']
df.drop(columns=drop_col, inplace= True)
df.to_csv('new_out_file.csv', index = False)
# print(df)
df.head()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

## procesude new technique

```python
X=[x for x in df.AvgTemp]
X
```

```python
size=int(len(X)*0.90)
train=X[0:size]
test=X[size:]
```

## end of experiemnts

```python
history=train
```

#discard the first column and set the index column 'Date'

```python
history
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
prediction=list()
for t in test:
    Arima=ARIMA(train,order=(1,0,5))
    model=Arima.fit()
    output=model.forecast()
    history.append(t)
    prediction.append(output)
    print('predicted=%f, expected=%f' % (output,t))
    
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rms = sqrt(mean_squared_error(test, prediction))
rms
```

```python
plt.plot(test,color='red')
plt.plot(prediction)
```

#best model---->(p,d,q)=(1,0,5)

```python
size  = int(len(df_v1)*0.90)
train = df_v1[:size]
test = df_v1[size:]
# print(len(train))
# print(len(test))

model = ARIMA(train['AvgTemp'], order=(2,0,5))
result = model.fit()
# result.summary()
```

#after fitting the model, predict the train data and plot the original and predict data:

```python
forecast_data = result.predict(start = len(train), end = len(df_v1)-1, dynamic=True)
#plot
plt.figure(figsize=(12, 5))
plt.plot(df_v1['AvgTemp'], label='Raw Data')
# plt.plot(train.index, train['AvgTemp'], label='Train Data', color='blue')
plt.plot(test.index, test['AvgTemp'], label='Test Data', color='orange')
plt.plot(test.index, forecast_data, label='Predict Data', color='green')
plt.title('ARIMA Model Forecast')
plt.xlabel('Date')
plt.ylabel('AvgTemp')
plt.legend()
plt.show()
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error


rmse = sqrt(mean_squared_error(forecast_data,test))

rmse
```

```python
from pandas import DataFrame


residual = DataFrame(result.resid)
residual.plot(kind='kde')
```
